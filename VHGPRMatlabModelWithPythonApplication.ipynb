This notebook shows how to run the matlab code using a python.For this we need the Matlab https://rutracker.org/forum/viewtopic.php?t=5490203. And also we need install MATLAB Engine API for Python (https://www.mathworks.com/help/matlab/matlab_external/install-the-matlab-engine-for-python.html). 
    MATLAB Engine API for Python is a free package that allows you to link python and matlab. 


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#include new package
import matlab.engine
```

After installation we start the engine

```python
eng = matlab.engine.start_matlab ()
```


Specify the path to the folder in which the matlab sources are located

```python
path = eng.addpath (r'./VHGPR/') 
```

So you can see the documentation for the function through matlab

```python
eng.doc("vhgpr_ui",nargout=0)
```

Load data and call matlab function 

```python
data = eng.load('motorcycle.mat')
y = data['y']
X = data['X']

NMSE, NLPD, Ey, Vy, mutst, diagSigmatst, atst, diagCtst, LambdaTheta, convergence = \
                                                                                eng.vhgpr_ui(X, y, X, y, 1, nargout=10)
```

Build graphics on our model

```python
Ey = np.array (Ey)
Vy = np.array (Vy)
X = np.array (X)

Ey.resize((133,))
Vy.resize((133,))
X.resize((133,))

Vy_ = 2 * np.sqrt (Vy)




plt.figure (figsize = (10, 7))
plt.scatter (X, y)
plt.plot (X, Ey, 'k', lw = 3, zorder = 9, label="predicted mean")
plt.fill_between (X, Ey - Vy_, Ey + Vy_, alpha=0.1, color='g')
```
